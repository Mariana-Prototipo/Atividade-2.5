1 - Projeto 1

Projete um sistema de memória que possui 13 pinos de endereços e um barramento de dados (BD) que transporta 2 palavras por vez. O BD não pode exceder 64 bits e a palavra deve ser maior que a célula da memória. A memória deve ter capacidade de 8 Gbits. Depois do projeto, indique qual é o tamanho da palavra, da célula, do BD e do barramento de endereços, e desenhe o chip.


```python
# ============================================================
# PROJETO DE MEMÓRIA
# 13 pinos de endereço
# Palavra = 32 bits
# Célula = 16 bits
# Barramento de Dados = 64 bits
# (2 palavras por acesso)
# Capacidade lógica = 8 Gbits
# ============================================================

class SistemaMemoria:

    def __init__(self):

        # Especificações do projeto
        self.barramento_endereco = 13
        self.total_enderecos = 2 ** self.barramento_endereco

        self.celula = 16          # bits
        self.palavra = 32         # bits
        self.bd = 64              # bits

        # Cada endereço possui uma palavra (32 bits)
        self.memoria = [0] * self.total_enderecos

    # -----------------------------------------------------
    # Escrever uma palavra de 32 bits
    # -----------------------------------------------------
    def escrever(self, endereco, valor):

        if endereco < 0 or endereco >= self.total_enderecos:
            print("Endereço inválido!")
            return

        # Mantém somente 32 bits
        self.memoria[endereco] = valor & 0xFFFFFFFF

        print(f"Palavra gravada no endereço {endereco}")

    # -----------------------------------------------------
    # Ler uma palavra
    # -----------------------------------------------------
    def ler(self, endereco):

        if endereco < 0 or endereco >= self.total_enderecos:
            print("Endereço inválido!")
            return

        valor = self.memoria[endereco]

        print(f"\nEndereço : {endereco}")
        print(f"Valor    : 0x{valor:08X}")

        return valor

    # -----------------------------------------------------
    # Ler duas palavras (64 bits)
    # -----------------------------------------------------
    def ler_duas_palavras(self, endereco):

        if endereco >= self.total_enderecos - 1:
            print("Não existem duas palavras disponíveis.")
            return

        p1 = self.memoria[endereco]
        p2 = self.memoria[endereco + 1]

        print("\n===== BARRAMENTO DE DADOS =====")
        print(f"Palavra 1 : {p1:032b}")
        print(f"Palavra 2 : {p2:032b}")

        barramento = (p1 << 32) | p2

        print(f"\nBD (64 bits):")
        print(f"{barramento:064b}")

    # -----------------------------------------------------
    # Informações do projeto
    # -----------------------------------------------------
    def especificacoes(self):

        print("\n=========== PROJETO ===========")
        print("Capacidade : 8 Gbits")
        print(f"Barramento de Endereços : {self.barramento_endereco} bits")
        print(f"Número de Endereços     : {self.total_enderecos}")
        print(f"Célula                  : {self.celula} bits")
        print(f"Palavra                 : {self.palavra} bits")
        print(f"Barramento de Dados     : {self.bd} bits")
        print("2 palavras por acesso")
        print("================================")

    # -----------------------------------------------------
    # Desenho do Chip
    # -----------------------------------------------------
    def desenhar_chip(self):

        print("""
                  CHIP DE MEMÓRIA

         +---------------------------------------+
A12 ---->|                                       |
A11 ---->|                                       |
A10 ---->|                                       |
A9  ---->|                                       |
A8  ---->|                                       |
A7  ---->|                                       |
A6  ---->|                                       |
A5  ---->|        MEMÓRIA 8 Gbits                |
A4  ---->|                                       |
A3  ---->| 8192 endereços                        |
A2  ---->| Palavra = 32 bits                     |
A1  ---->| Célula = 16 bits                      |
A0  ---->|                                       |
         |                                       |
CS ----->|                                       |
OE ----->|                                       |
WE ----->|                                       |
         |                                       |
D63 <--->|                                       |
...      |                                       |
D0  <--->|                                       |
         +---------------------------------------+

Barramento de Endereços = 13 bits
Barramento de Dados = 64 bits
""")


# ============================================================
# Programa Principal
# ============================================================

def main():

    memoria = SistemaMemoria()

    while True:

        print("\n========== MENU ==========")
        print("1 - Escrever Palavra")
        print("2 - Ler Palavra")
        print("3 - Ler 2 Palavras (BD)")
        print("4 - Especificações")
        print("5 - Desenhar Chip")
        print("6 - Sair")

        op = input("Escolha: ")

        if op == "1":

            endereco = int(input("Endereço (0 a 8191): "))
            valor = int(input("Valor hexadecimal: "), 16)

            memoria.escrever(endereco, valor)

        elif op == "2":

            endereco = int(input("Endereço: "))
            memoria.ler(endereco)

        elif op == "3":

            endereco = int(input("Endereço inicial: "))
            memoria.ler_duas_palavras(endereco)

        elif op == "4":

            memoria.especificacoes()

        elif op == "5":

            memoria.desenhar_chip()

        elif op == "6":

            print("Programa encerrado.")
            break

        else:

            print("Opção inválida.")


if __name__ == "__main__":
    main()
```

2 - Projeto 2

Projete um sistema de memória que possui 16 linhas de dados (BD). Os pinos de endereço precisam ser menores que 14. A memória deve ter capacidade de 32 Gbits. Justifique cada decisão do projeto. Desenhe o chip. Qual é o tamanho do barramento de endereços?


```python
# ==========================================================
# SIMULADOR DE MEMÓRIA
# Projeto:
# Barramento de Dados (BD): 16 bits
# Barramento de Endereços: 13 bits
# Número de Endereços: 8192
# Capacidade lógica: 32 Gbits
# ==========================================================

class ChipMemoria:

    def __init__(self):

        # Quantidade de linhas do barramento de endereços
        self.barramento_endereco = 13

        # Número de endereços da memória
        self.total_enderecos = 2 ** self.barramento_endereco

        # Barramento de dados (16 bits)
        self.bd = 16

        # Capacidade lógica da memória
        self.capacidade = "32 Gbits"

        # Cria a memória com 8192 posições
        # Cada posição armazena um valor de 16 bits
        self.memoria = [0] * self.total_enderecos

    # -------------------------------------------------------
    # Escreve um valor em um endereço
    # -------------------------------------------------------
    def escrever(self, endereco, valor):

        # Verifica se o endereço existe
        if endereco < 0 or endereco >= self.total_enderecos:
            print("Endereço inválido!")
            return

        # Mantém somente 16 bits
        self.memoria[endereco] = valor & 0xFFFF

        print(f"Valor {valor:04X} gravado no endereço {endereco}")

    # -------------------------------------------------------
    # Lê um endereço
    # -------------------------------------------------------
    def ler(self, endereco):

        if endereco < 0 or endereco >= self.total_enderecos:
            print("Endereço inválido!")
            return

        valor = self.memoria[endereco]

        print(f"Endereço {endereco}")
        print(f"Valor = {valor:04X}")

        return valor

    # -------------------------------------------------------
    # Mostra informações do projeto
    # -------------------------------------------------------
    def informacoes(self):

        print("\n===== ESPECIFICAÇÕES =====")
        print(f"Capacidade da Memória : {self.capacidade}")
        print(f"Barramento de Dados   : {self.bd} bits")
        print(f"Barramento Endereço   : {self.barramento_endereco} bits")
        print(f"Número de Endereços   : {self.total_enderecos}")
        print("==========================")

    # -------------------------------------------------------
    # Desenho do chip
    # -------------------------------------------------------
    def desenho_chip(self):

        print("""
                 CHIP DE MEMÓRIA

          +---------------------------+
A12 ----->|                           |
A11 ----->|                           |
A10 ----->|                           |
A9  ----->|                           |
A8  ----->|                           |
A7  ----->|                           |
A6  ----->|                           |
A5  ----->|      MEMÓRIA 32 Gbits     |
A4  ----->|                           |
A3  ----->|      8192 Endereços       |
A2  ----->|                           |
A1  ----->|                           |
A0  ----->|                           |
          |                           |
CS ------>|                           |
OE ------>|                           |
WE ------>|                           |
          |                           |
D15 <-----|                           |
...       |                           |
D0  <-----|                           |
          +---------------------------+

Barramento de Endereços = 13 bits
Barramento de Dados = 16 bits
""")


# ==========================================================
# Programa Principal
# ==========================================================

def main():

    chip = ChipMemoria()

    while True:

        print("\n========== MENU ==========")
        print("1 - Escrever")
        print("2 - Ler")
        print("3 - Informações")
        print("4 - Desenhar Chip")
        print("5 - Sair")

        opcao = input("Escolha: ")

        if opcao == "1":

            endereco = int(input("Endereço (0-8191): "))
            valor = int(input("Valor hexadecimal: "), 16)

            chip.escrever(endereco, valor)

        elif opcao == "2":

            endereco = int(input("Endereço: "))
            chip.ler(endereco)

        elif opcao == "3":

            chip.informacoes()

        elif opcao == "4":

            chip.desenho_chip()

        elif opcao == "5":

            print("Programa encerrado.")
            break

        else:

            print("Opção inválida.")


if __name__ == "__main__":
    main()
```
